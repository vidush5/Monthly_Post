```python
#Remove empty and single valued columns as well as empty, duplicates
df1 = klib.data_cleaning(df)
df1
```

# Correlation Plot

```python
# Display all correlations data
klib.corr_plot(df1, annot=False)
```

```python
# Display only positive correlation
klib.corr_plot(df1,split='pos', annot=False)
```

```python
#Display only negative correlation
klib.corr_plot(df1,split='neg', annot=False)
```

# Save Plot as a PDF file
